** Import pandas and read in the Ecommerce Purchases csv file and set it to a DataFrame called ecom. **

```python
# import pandas
import pandas as pd
```

```python
# Read in the Ecommerce Purchases dataset and set it to a DataFrame called 'ecom'
ecom = pd.read_csv('Ecommerce_Purchases.csv')
```

**Check the head of the DataFrame.**

```python
# check the head of the DataFrame
ecom.head() # ecom.head() shows the first 5 rows of the DataFrame
```

** How many rows and columns are there? **

```python
# NO. OF ROWS AND COLUMNS
# How many rows and columns are there?
ecom.info()
```

** What is the average Purchase Price? **

```python
# mean of Purchase Price

ecom['Purchase Price'].mean()
```

** What were the highest and lowest purchase prices? **

```python
# the highest purchase price
ecom['Purchase Price'].max()
```

```python
# the lowest purchase price
ecom['Purchase Price'].min()
```

** How many people have English 'en' as their Language of choice on the website? **

```python
ecom[ecom['Language']== 'en']
```

** How many people have the job title of "Lawyer" ? **


```python
ecom[ecom['Job']== 'Lawyer']
```

** How many people made the purchase during the AM and how many people made the purchase during PM ? **

**(Hint: Check out [value_counts()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html) ) **

```python
# how many people made the purchase during AM time and how many during PM time?

ecom['AM or PM'].value_counts() # .value_counts() gives the count of unique values in the column
```

** What are the 5 most common Job Titles? **

```python
# 5 most common Job Titles
ecom['Job'].value_counts().head(5) # .head(5) gives the first 5 rows of the Series  
```

** Someone made a purchase that came from Lot: "90 WT" , what was the Purchase Price for this transaction? **

```python
# purchase price for Lot '90 WT'

ecom[ecom['Lot']=='90 WT']['Purchase Price']
```

** What is the email of the person with the following Credit Card Number: 4926535242672853 **

```python
# email of the person with credit card number '4926535242672853'

ecom[ecom['Credit Card']== 4926535242672853]['Email']
```

** How many people have American Express as their Credit Card Provider *and* made a purchase above $95 ?**

```python
ecom[(ecom['CC Provider']=='American Express') & (ecom['Purchase Price'] > 95)].value_counts() # .value_counts() gives the count of unique values in the column
```

** Hard: How many people have a credit card that expires in 2025? **

```python
# How many people have a credit card that expires in 2025?
ecom['CC Exp Date'].apply(lambda x: x.split('/')[1]).value_counts().get('25', 0)  # Count of '25' in the year part of the expiration date
```

** Hard: What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...) **

```python
def most_popular_emailproviders(emails):
    return emails.split('@')[1] #splits the email at '@' and returns the second part (the domain)

ecom['Email'].apply(most_popular_emailproviders).value_counts().head(5)  # .head(5) gives the first 5 rows of the Series
    
```
